```python
import os

import pandas as pd

gt_filename = "../../data/31_c.ply"

directory = '/media/linsel/3931-6632/data/phd stone tools/Blades_Bladelets_PLY/results/'

paths = [x[0] for x in os.walk(directory)][1:]

files = [x[0].split('/')[len(x[0].split('/'))-1].split('_')[0] + '_result_table.csv' for x in os.walk(directory)][1:]

files_list = [x[0].split('/')[len(x[0].split('/'))-1].split('_')[0]  for x in os.walk(directory)][1:]

result_list = []

ids = []

for num,path in enumerate(paths):
    
    data_list = pd.read_csv(path + '/' + files[num],sep=',', index_col=None,header=None)
    
    result_list.append(data_list)

    ids.extend([files_list[num]]*len(data_list))

    print(len(ids))
```

```python
frame = pd.concat(result_list, axis=0, ignore_index=True)
             
frame = frame.rename(columns={0: 'Persistence', 1: 'High-Pass',2: 'Low-Pass', 3: 'Merge',4: 'Res',5:'ID'})

frame['ID'] = ids
```

```python
frame['Persistence'] = np.round_(frame['Persistence'],2)
frame['Low-Pass'] = np.round_(frame['Low-Pass'],2)
frame['High-Pass'] = np.round_(frame['High-Pass'],2)
frame['Merge'] = np.round_(frame['Merge'],2)
frame['Res'] = np.round_(frame['Res'],0)
```
