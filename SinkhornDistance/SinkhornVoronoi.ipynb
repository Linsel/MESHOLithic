```python
import os
import numpy as np
from sinkhornVoronoi import sinkhornvoronoi

path = '/media/linsel/3931-6632/data/phd stone tools/Blades_Bladelets_PLY/annotated/coloured/selected/'
exp_path = 'exp_path/'

i = 0
files = []
num = 1000

for n,file in enumerate(os.listdir(path)):
    if file.endswith('.ply'):
        files.append(file[:-4])
        i += 1

dict_kmeans = {}

for n,name in enumerate(files):

    sv = sinkhornvoronoi()
    sv.load_ply(path ,name,exp_path)
    sv.sample_voronoi(num)  

    dict_kmeans[name] = sv.centroids_nearest_neighbor
```

```python
import ot

# calculate the sinkhorn distance
def calc_unbalanced_OT(a,b):

    # Compute the cost matrix between each pair of points in the two meshes
    C = np.sum((a[:, None, :] - b[None, :, :]) ** 2, axis=-1)

    # Compute the Sinkhorn distance
    reg = 1e-3
    G0 = ot.sinkhorn_unbalanced(np.ones(n) / n, np.ones(n) / n, M = C, reg = reg, method='sinkhorn')
    sinkhorn_dist = np.sum(G0 * C)
    return sinkhorn_dist

def calc_EMD(a,b,n):

    # Compute the cost matrix between each pair of points in the two meshes
    C = np.sum((a[:, None, :] - b[None, :, :]) ** 2, axis=-1)

    # Compute the Sinkhorn distance
    reg = 1e-3
    G0 = ot.emd(np.ones(n) / n, np.ones(n) / n, C, reg)
    sinkhorn_dist = np.sum(G0 * C)
    return sinkhorn_dist
```
